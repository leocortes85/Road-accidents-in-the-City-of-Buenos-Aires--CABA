In order to join this dataframe with the facts dataframe, we eliminate the columns that are repeated in both. We will also eliminate the "FECHA_FALLECIMIENTO" column since it does not provide relevant information for the analysis we must perform.

```python
columns_to_drop = ['FECHA', 'AAAA', 'MM', 'DD', 'VICTIMA', 'FECHA_FALLECIMIENTO']

# Eliminar las columnas
victims = victims.drop(columns_to_drop, axis=1)
```

```python
victims.columns = victims.columns.str.replace('_', ' ')

victims.columns = victims.columns.str.title()
# Rename columns
victims = victims.rename(columns={'Id Hecho': 'Hechos_Id'})
victims.columns
```

```python
# Calculate the mean and standard deviation of the column 'Edad' (ignoring the 'SD' values)
avg_age = victims[victims['Edad'] != 'SD']['Edad'].astype(float).mean()
std_dev = victims[victims['Edad'] != 'SD']['Edad'].astype(float).std()

# Generate random numbers within the defined range
np.random.seed(42)  # This ensures that the random numbers are reproducible
random_age = np.random.randint(
    avg_age - 2 * std_dev,
    avg_age + 2 * std_dev + 1,  # +1 to include the upper limit
    size=len(victims[victims['Edad'] == 'SD'])
)

# Replace 'SD' values ​​with the generated random numbers
victims.loc[victims['Edad'] == 'SD', 'Edad'] = random_age

# Convert column 'Edad' back to integer type
victims['Edad'] = victims['Edad'].astype(int)

victims['Edad'].unique()
```
